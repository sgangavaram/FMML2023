```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
data.info()
```

```python
data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(24,7)})
sns.boxplot(x=X["make"],y=y, palette="Set3").set_title('Car Manufacturer vs Price - Box Plot')
```

```python
sns.violinplot(x=X["make"],y=y, palette="Set3").set_title('Car maker vs Price - Violin Plot')
```

```python
sns.swarmplot(x=X["make"],y=y).set_title('Car maker vs Price Swarm Plot')
```

```python
sns.boxplot(x=X["make"],y=pd.to_numeric(X["horsepower"]), palette="Set3").set_title('Car maker vs Horsepower Box Plot')
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.scatterplot(x=pd.to_numeric(X["horsepower"]), y = y, color = "orange")
```

```python
sns.jointplot(x=pd.to_numeric(X["horsepower"]), y = y, kind="reg", color = 'red')
```

**1) Think about the maximum/minimum values for each body-type, which body-type is the "costliest" ?**

Upon examining the highest and lowest price points associated with each body type, it becomes evident that sedans rank as the most expensive among them.

**2) Do you think maximum/minimum values are a clear indication of which body-type is costliest, or does the interpretation improve by also taking the interquartile range into consideration?**

Looking at just the highest and lowest values may not give the full picture of the most expensive body type, as extreme values can skew the results. The interquartile range (IQR) is a useful measure that looks at the middle 50% of the data, giving a better idea of how costs are spread out.

When figuring out the priciest body type, it's important to look at both the average (mean or median) and how the costs vary (range, standard deviation, or IQR). The mean can be affected by outliers, so if there are extreme values, using the median might be better. Including the IQR helps pinpoint where most of the data lies, making it less influenced by extreme values.

To sum it up, while the highest and lowest values show the data range, adding measures like the interquartile range gives a more detailed view of the distribution, helping to better understand which body type is the most expensive.


**3) Do you think the body-style of a car can be a good predictor for its price?**

The type of car design can affect its price, but it's just one of many factors. Different things determine how much a car costs, like the brand, model, features, how well it performs, and how popular it is.

Fancy cars like luxury sedans or high-performance sports cars usually cost more because they use high-quality materials and have advanced technology. On the other hand, smaller cars or budget models are often cheaper but still give good value for what they're meant for.

When figuring out the price of a car, it's important to look at the features it has, the reputation of the brand, and how well it's made, not just its design. Other things, like how many people want to buy it and what's happening in the economy, can also affect the price. So, while the type of design can give you an idea, it's not the only thing that decides how much a car costs.
